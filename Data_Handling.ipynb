```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset("titanic")
df = df[["survived", "sex", "age", "fare", "class", "embarked"]].dropna()
print("Dataset loaded successfully  \n")
```

```python
plt.figure(figsize = (8,5))
sns.countplot(data = df, x = "sex", hue = "survived", palette = "Set2")
plt.title("Survival Count by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title = "Survived" , labels = ["No", "Yes"])
plt.show()

plt.figure(figsize = (8,5))
sns.lineplot(data = df.sort_values("age"), x = "age", y = "fare", color = "purple")
plt.title("Fare Trend with Increasing Age")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()

plt.figure(figsize = (7,5))
sns.scatterplot(data = df, x = "age", y = "fare", hue = "sex", palette = "coolwarm")
plt.title("Age vs Fare by Gender")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
df["class"].unique()
df["class"].value_counts()
```

```python
plt.figure(figsize = (7 , 7))
df["class"].value_counts().plot(kind = "pie", autopct = "%1.1f%%" ,colors = ["Yellow", "Blue", "Pink"])
plt.title("Distribution of Passenger Class")
plt.ylabel("")
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
sns.histplot(df["age"], bins = 20, kde = True, color = "skyblue", ax = axes[0,0])
axes[0,0].set_title("Age Distribution")
sns.boxplot(data = df, x = "fare", color = "orange", ax = axes[0 , 1])
axes[0,1].set_title("Fare Boxplot")
sns.countplot(data = df, x = "survived", palette = "Set1" , ax = axes[1 , 0])
axes[1,0].set_title("Survival Count")
sns.scatterplot(data = df, x = "age", y = "fare", color = "green", ax = axes[1,1])
axes[1,1].set_title("Age Vs Fare")
plt.tight_layout()
plt.show()
```
